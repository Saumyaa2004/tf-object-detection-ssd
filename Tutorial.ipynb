```python
import tensorflow as tf
import cv2
import numpy as np
from IPython.display import display, clear_output
from PIL import Image
from object_detection.utils import label_map_util, visualization_utils as viz_utils
from object_detection.utils import config_util
from object_detection.builders import model_builder


# Fix for deprecated tf.gfile in TF2
tf.gfile = tf.io.gfile

# Paths
ANNOTATION_PATH = 'C:/RealTimeObjectDetection/Tensorflow/workspace/annotations'
CHECKPOINT_PATH = 'C:/RealTimeObjectDetection/Tensorflow/workspace/models/my_ssd_mobnet'
CONFIG_PATH = 'C:/RealTimeObjectDetection/Tensorflow/workspace/models/my_ssd_mobnet/pipeline.config'

# Load category index
category_index = label_map_util.create_category_index_from_labelmap(
    ANNOTATION_PATH + '/label_map.pbtxt'
)

# Load pipeline config and build detection model
configs = config_util.get_configs_from_pipeline_file(CONFIG_PATH)
detection_model = model_builder.build(model_config=configs['model'], is_training=False)

# Restore latest checkpoint
latest_ckpt = tf.train.latest_checkpoint(CHECKPOINT_PATH)
ckpt = tf.compat.v2.train.Checkpoint(model=detection_model)
ckpt.restore(latest_ckpt).expect_partial()
print("Restored checkpoint:", latest_ckpt)

# Detection function
@tf.function
def detect_fn(image):
    image, shapes = detection_model.preprocess(image)
    prediction_dict = detection_model.predict(image, shapes)
    detections = detection_model.postprocess(prediction_dict, shapes)
    return detections

# Initialize webcam
cap = cv2.VideoCapture(0)

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Prepare input
        image_np = np.array(frame)
        input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0), dtype=tf.float32)

        # Run detection
        detections = detect_fn(input_tensor)
        num_detections = int(detections.pop('num_detections'))
        detections = {key: value[0, :num_detections].numpy()
                      for key, value in detections.items()}
        detections['num_detections'] = num_detections
        detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

        # Visualize results
        image_np_with_detections = image_np.copy()
        viz_utils.visualize_boxes_and_labels_on_image_array(
            image_np_with_detections,
            detections['detection_boxes'],
            detections['detection_classes'] + 1,
            detections['detection_scores'],
            category_index,
            use_normalized_coordinates=True,
            max_boxes_to_draw=5,
            min_score_thresh=0.5,
            agnostic_mode=False
        )

        # Display in notebook
        img = Image.fromarray(cv2.cvtColor(image_np_with_detections, cv2.COLOR_BGR2RGB))
        clear_output(wait=True)
        display(img)

except KeyboardInterrupt:
    print("Stopped webcam")
finally:
    cap.release()
```
